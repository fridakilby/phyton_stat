### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# Plotting can be done in two ways either by oo-style or by pyplot. (oo = object-oriented)
# The one recomended is by oo-style since you call every method by it self and don't rely on pyplot to create your Figures and Axes
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# You could se a figure as a whole canvas.
# Inside the figure you can put one or so many Axes as you will, which you fill with data.
# Axis is x, y or more axis that you can name to whatever you like, they show the intervall for values.
# Artist is everything you see on the figure. Each element in represented by an artist.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
# Expected input data-type in matplotlib are an array from numpy or an array-like (masked) input.
# If your data isn't an array you can convert it by using the function in numpy asarray
# For example: you can make a list to a numpy array by using asarray
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# with oo-style
x = np.linspace(-4, 4, 100) # Sampling data from -4 to 4, 100 values
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained') # creating figure and axes
ax.plot(x, x**2, label='my label') # creating a artist
ax.set_xlabel('x-axel') # labeling the x-axis
ax.set_ylabel('y-axel') # labeling the y-axis
ax.set_title('oo-style') # setting title 
ax.legend() # wrinting the label
```

```python
# pyplot-style
y = np.linspace(-4, 4, 100)
plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(y, y**2, label='my label')
plt.xlabel('x-axel')
plt.ylabel('y-axel')
plt.title('pyplot-style')
plt.legend()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, axs= plt.subplots(2, 1, figsize=(5, 5), layout='constrained')

axs[0].scatter(random_data_x, random_data_y, label='scatter', color='green') # choosed to use random_Data
axs[0].set_title('Scatter-plot')
# axs[0].legend()

axs[1].bar(names, values, label=('bar'), color='yellow')
axs[1].set_title('Bar-plot')
axs[1].set_xlabel('Fruits')
axs[1].set_ylabel('Amount')
# axs[1].legend()
```
